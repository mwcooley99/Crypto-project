# Crypto Data


```python
# Import dependencies
import numpy as np
import pandas as pd
from sqlalchemy import create_engine
import datetime as dt
```

```python
# Importing csv into dataframe
csv_file = "./Resources/crypto-markets.csv"
crypto_data = pd.read_csv(csv_file, encoding='cp1252')
crypto_data.head(5)
```

```python
# CRYPTO

# Split column to extract a Year
crypto_data['date'] = pd.to_datetime(crypto_data.date)
crypto_data['year'] = crypto_data.date.dt.year

# Choose the relevant columns
crypto_data = crypto_data[['name', 'symbol', 'date', 'year', 'open', 'high', 'low', 'close', 'volume', 'market', 'close_ratio', 'spread']]

# Choose the relevant data
crypto_data = crypto_data[crypto_data['name'].isin(['Bitcoin', 'Ethereum','Litecoin', 'Ripple', 'Bitcoin Cash', 'EOS', 'Binance', 'Bitcoin SV', 'Stellar', 'Cardano', 'Dash', 'Monero'])]
crypto_data = crypto_data.reset_index()
crypto_data.head()
```

```python
# BITCOIN

bitcoin_data = crypto_data.loc[(crypto_data["name"] == 'Bitcoin'),:]
bitcoin_data.head()
```

```python
# ETHEREUM

ethereum_data = crypto_data.loc[(crypto_data["name"] == 'Ethereum'),:]
ethereum_data.head()
```
